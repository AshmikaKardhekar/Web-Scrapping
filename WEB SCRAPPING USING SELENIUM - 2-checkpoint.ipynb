# Question 2 :-
    In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: 
    "Brand Name",
    "Name of the Product", 
    "Rating", 
    "No. of Ratings", 
    "Price", 
    "Return/Exchange", 
    "Expected Delivery", 
    "Availability",
    "Other Details" 
    and “Product URL”.
    In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Creating the empty lists wherein we will store the data related to that particular title
Brand_Name = []
Product_Name = []
Rating = []
No_of_ratings = []
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
Other_details = []
Product_URL = []
```

```python
# Navigate through the pages until 3rd page and if not present scrape the available data
for page in range(0,3):
    
    # For clicking on the product link
    pdt_link = driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']//span[@class='a-size-medium a-color-base a-text-normal']")
    for p in pdt_link:
        page_end = 0
        p.click()
        time.sleep(5)

        # For switching to new window
        driver.switch_to.window(driver.window_handles[-1])
    
        # Finding the tags for Product Name 
        try:
            Product_Name.append((driver.find_element_by_xpath("//span[@id='productTitle' and @class='a-size-large product-title-word-break']")).text)
        except:
            Product_Name.append("-")
        
        # Finding the tags for Brand Name 
        try:
            Brand_Name.append((driver.find_element_by_xpath("//table[@id='productDetails_techSpec_section_1' and @class='a-keyvalue prodDetTable']//tr[1]//td[@class='a-size-base prodDetAttrValue']")).text)
        except:
            Brand_Name.append("-")

        # Finding the tags for Rating 
        try:
            Rating.append((driver.find_element_by_xpath("//span[@id='acrPopover' and @class='reviewCountTextLinkedHistogram noUnderline']")).get_attribute('title'))
        except:
            Rating.append("-")
        
        # Finding the tags for No_of_ratings 
        try:
            No_of_ratings.append((driver.find_element_by_xpath("//span[@id='acrCustomerReviewText' and @class='a-size-base']")).text)
        except:
            No_of_ratings.append("-")
        
        # Finding the tags for Price
        try:
            Price.append((driver.find_element_by_xpath("//span[@id='priceblock_dealprice' or @id='priceblock_ourprice']")).text)
        except:
            Price.append("-")

        # Finding the tags for Return_Exchange
        try:
            Return_Exchange.append((driver.find_element_by_xpath(".//div[@id='RETURNS_POLICY' and @class='a-column a-span3 a-text-center icon-container  icon-grid-width celwidget']")).text)
        except:
            Return_Exchange.append("-")

        # Finding the tags for Expected_Delivery
        try:
            Expected_Delivery.append((driver.find_element_by_xpath(".//div[@id='ddmDeliveryMessage' and @class='a-section a-spacing-mini']//b")).text)
        except:
            Expected_Delivery.append("-")

        # Finding the tags for Availability
        try:
            Availability.append((driver.find_element_by_xpath(".//div[@id='availability' and @class='a-section a-spacing-none']/span[@class='a-size-medium a-color-success']")).text)
        except:
            Availability.append("-")

        # Finding the tags for Other_details
        try:
            Other_details.append((driver.find_element_by_xpath(".//div[@id='feature-bullets' and @class='a-section a-spacing-medium a-spacing-top-small']/ul[@class='a-unordered-list a-vertical a-spacing-mini']")).text)
        except:
            Other_details.append("-")

        # Finding the tags for Product_URL
        Product_URL.append(driver.current_url)

        # For switching to the first window
        driver.switch_to.window(driver.window_handles[0])
    
        time.sleep(5)
    
    # Finding the tags for clicking on the next button
    try:
        driver.find_element_by_xpath("//ul[@class='a-pagination']//li[@class='a-last']/a[text()='Next']").click()
    except:
        page_end = 2
    # Introducing a temporary variable to create an error so that the loop breaks are scraps the information for the previous pages    
    if page_end == 2:
        break
        
    time.sleep(5)
```

```python
#Building the data frame
data = pd.DataFrame({
        'Product_Name' : Product_Name,
        'Brand_Name' : Brand_Name,
        'Rating' : Rating,
        'No_of_ratings' : No_of_ratings,
        'Price' : Price,
        'Return_Exchange' : Return_Exchange,
        'Expected_Delivery' : Expected_Delivery,
        'Availability' : Availability,
        'Other_details' : Other_details,
        'Product_URL' : Product_URL
})
    
print('details ----->')
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', None)
# Saving the DataFrame to CSV
data.to_csv('details.csv')
display(data)
```

```python
name = []
url = []
search = ['fruits', 'cars', 'Machine Learning']
for s in search:
    
    # Getting the tags for search bar & search button
    search_bar = driver.find_element_by_xpath("//input[@class='gLFyf gsfi' or @class='JSAgYe']")
    
    search_bar.clear()
    
    search_bar.send_keys(s)

    search_button = driver.find_element_by_xpath("//button[@class='Tg7LZd' or @class='KXJfe']")
    search_button.click()
    
    time.sleep(5)

    # Adding the code to scroll the page
    driver.execute_script("window.scrollTo(0,5000);")
    
    time.sleep(5)
    
    # Scrapping the images from the fruits tags
    url_tags = driver.find_elements_by_xpath("//a[@class='wXeWr islib nfEiy mM5pbd']//div[@class='bRMDJf islir']")
    for u in url_tags[0:100]:
        driver.execute_script("arguments[0].scrollIntoView();", u)
        time.sleep(2)
        url.append(u.find_element_by_xpath(".//img[@class='rg_i Q4LuWd']").get_attribute('src'))
        name.append(s)
         
    time.sleep(5)
    
```

# Question 6 :-
    Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
# Writing a request to fetch the data from the web page 
driver.get("https://trak.in/")
driver.maximize_window()
```

```python
# Finding the tag for clicking the section of 'Funding Deals'
fd = driver.find_element_by_xpath("//a[@title='https://trak.in/india-startup-funding-investment-2015/']")
fd.click()
```

```python
# Clicking on the maximun number of entries available
click_100 = driver.find_elements_by_xpath("//*[text()='100']")
for c in click_100:
    c.click()
```

```python
# Creating empty lists for storing the relevent data
Date = []
Startup_Name = []
Industry_Vertical = []
Sub_Vertical = []
City_Location = []
Investors_Name = []
Investment_Type = []
Amount = []
```

```python
# Finding the tag for header elements
year_2020 = driver.find_elements_by_xpath("//h2")

for j in year_2020:
    
    # Getting the tags for the months whose details are to be scrapped
    if j.text in ["July, 2020", "August, 2020", "September, 2020"]:
        row = j.find_element_by_xpath("./following-sibling::div[@class='dataTables_wrapper no-footer']//tbody[@class='row-hover']")
        
        for r in row.find_elements_by_xpath(".//tr[@role='row']"):
           
            # Finding tags for Date
            Date_tags = r.find_elements_by_xpath(".//td[@class='column-2']")
            for d in Date_tags:
                Date.append(d.text)
                
            # Finding tags for Startup's name
            Name_tags = r.find_elements_by_xpath(".//td[@class='column-3']")
            for n in Name_tags:
                Startup_Name.append(n.text)
                
            # Finding tags for Industry/Vertical name
            Vertical_tags = r.find_elements_by_xpath(".//td[@class='column-4']")
            for v in Vertical_tags:
                Industry_Vertical.append(v.text)
                
            # Finding tags for Sub Vertical name
            Sub_tags = r.find_elements_by_xpath(".//td[@class='column-5']")
            for s in Sub_tags:
                Sub_Vertical.append(s.text)
                
            # Finding tags for Location name
            Location_tags = r.find_elements_by_xpath(".//td[@class='column-6']")
            for l in Location_tags:
                City_Location.append(l.text)
                
            # Finding tags for Investors name 
            Investor_tags = r.find_elements_by_xpath(".//td[@class='column-7']")
            for i in Investor_tags:
                Investors_Name.append(i.text)
                
            # Finding tags for Investment type
            Investment_tags = r.find_elements_by_xpath(".//td[@class='column-8']")
            for it in Investment_tags:
                Investment_Type.append(it.text)
            
            # Finding tags for Amount
            Amount_tags = r.find_elements_by_xpath(".//td[@class='column-9']")
            for a in Amount_tags:
                Amount.append(a.text)
        
```

```python
#Building the data frame
funding_deals_data = pd.DataFrame({
        'Date (dd/mm/yyyy)' : Date,
        'Startup_Name' : Startup_Name,
        'Industry/Vertical' : Industry_Vertical,
        'Sub_Vertical' : Sub_Vertical,
        'City_Location' : City_Location,
        'Investors_Name' : Investors_Name,
        'Investment_Type' : Investment_Type,
        'Amount (In USD)' : Amount
})
    
print('Details of all the funding deals for second quarter (i.e. July 20 – September 20) ----->')
pd.set_option('display.max_rows', 40)
pd.set_option('display.max_colwidth', None)
funding_deals_data
```

# Question 7 :-
    Write a program to scrap all the available details of top 10 gaming laptops from digit.in. 

```python
# Writing a request to fetch the data from the web page 
driver.get("https://www.digit.in/")
```

```python
# Finding the clickable tag for the top 10 gaming laptop
driver.find_element_by_xpath("//a[@href='https://www.digit.in/top-products/best-gaming-laptops-40.html']").click()
```

```python
# Creating empty lists to save the corresponding extracted information
Product = []
OS = []
Display = []
Processor = []
Memory = []
Weight = []
Dimension = []
Graphics_Processor = []
Seller = []
Price = []

best_seller = driver.find_elements_by_xpath("//div[@class='Top10-Seller']")
table = driver.find_elements_by_xpath("//div[@class='Spcs-details']")

for b in best_seller:
    
    # Finding the tags for the product name
    product_tags = b.find_elements_by_xpath(".//tbody//td[1]")
    for p in product_tags:
        Product.append(p.text)
        
    # Finding the tags for seller name
    seller_tags = b.find_elements_by_xpath(".//td[@class='smmerchant']")
    for s in seller_tags:
        Seller.append(s.text)  
            
    # Finding the tags for price
    price_tags = b.find_elements_by_xpath(".//td[@class='smprice']")
    for pr in price_tags:
        Price.append(pr.text)  
        
for t in table:

    # Finding tags for operating system
    os_tags = t.find_elements_by_xpath(".//tr[3]//td[3]")
    for o in os_tags:
        OS.append(o.text)
        
    # Finding tags for display
    display_tags = t.find_elements_by_xpath(".//tr[4]//td[3]")
    for d in display_tags:
        Display.append(d.text)
        
    # Finding tags for processor
    processor_tags = t.find_elements_by_xpath(".//tr[5]//td[3]")
    for pc in processor_tags:
        Processor.append(pc.text)
     
    # Finding tags for memory
    memory_tags = t.find_elements_by_xpath(".//tr[6]//td[3]")
    for m in memory_tags:
        Memory.append(m.text)
        
    # Finding tags forweight
    weight_tags = t.find_elements_by_xpath(".//tr[7]//td[3]")
    for w in weight_tags:
        Weight.append(w.text)
        
    # Finding tags for dimension
    dimension_tags = t.find_elements_by_xpath(".//tr[8]//td[3]")
    for d in dimension_tags:
        Dimension.append(d.text)
        
    # Finding tags for operating system
    gp_tags = t.find_elements_by_xpath(".//tr[9]//td[3]")
    for g in gp_tags:
        Graphics_Processor.append(g.text)
    
    
```

```python
#Building the data frame
gaming_laptops_data = pd.DataFrame({
        'Product' : Product,
        'Seller' : Seller,
        'Price' : Price,
        'OS' : OS,
        'Dsiplay' : Display,
        'Processor' : Processor,
        'Memory' : Memory,
        'Weight' : Weight,
        'Dimension' : Dimension,
        'Graphics_processor' : Graphics_Processor
})
    
print('Available details of top 10 gaming laptops ----->')
pd.set_option('display.max_rows', 40)
pd.set_option('display.max_colwidth', None)
gaming_laptops_data
```
