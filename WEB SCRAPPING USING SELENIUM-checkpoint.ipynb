```python
# Scrapping the job description required from the page
job_description_tags = driver.find_elements_by_xpath("//div[@class='info fleft']/a")
job_description = []
for jd in job_description_tags:
    
    try:
        jd.click()
    except:
        
        try:
            ref = jd.get_attribute("href")
            path = "//div[@class="'info fleft'"]//a[@href'"+ref+"']"
            # Finding element for the click button
            element = WebDriverWait(driver,10).until(
            EC.element_to_be_clickable((By.XPATH, path))
            )
            element.click()
        except:
            job_description.append('Exception Error!')
    
    time.sleep(5)
    
    driver.switch_to.window(driver.window_handles[-1])
    
    try:
        j = driver.find_element_by_xpath('//section[@class="job-desc"]')
        job_description.append(j.text)
    except:
        job_description.append('Not found on the website of naukri.com!')

    driver.switch_to.window(driver.window_handles[0])
    
    time.sleep(5)
```

```python
print(len(job_description))
```

```python
#Building the data frame
job_data = pd.DataFrame({
        'job_title' : job_title[0:10],
        'job_location' : job_location[0:10],
        'company_name' : company_name[0:10],
        'job_description' : job_description[0:10]
})
    
print('First 10 jobs data. ----->')
pd.set_option('display.max_colwidth', None)
display(job_data)
```

# Question 4 :-
    Write a python program to scrape data for first 10 job results for Data scientist Designation in Noida location. You have to scrape company_name, No. of days ago when job was posted, Rating of the company.

```python
# Finding  element for the job search bar
search_job = driver.find_element_by_xpath("//input[@id='KeywordSearch']")
search_job.clear()
search_job.send_keys('Data Scientist')

# Finding element for the job location
search_location = driver.find_element_by_xpath("//input[@id='LocationSearch']")
search_location.clear()
search_location.send_keys('Noida')
```

```python
# Scrapping for company name from the page
company_name_tags = driver.find_elements_by_xpath("//div[@class='d-flex justify-content-between align-items-start']")
company_name = [c.text for c in company_name_tags]
print(len(company_name))
print(company_name)

# Scrapping for no of days posted before
days_posted_ago_tags = driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']")
days_posted_ago = [a.text for a in days_posted_ago_tags]
print(len(days_posted_ago))
print(days_posted_ago)

# Scrapping for the ratings
ratings_tags = driver.find_elements_by_xpath("//div[@class='d-flex flex-column css-x75kgh e1rrn5ka3']//span[@class='css-19pjha7 e1cjmv6j1']")
ratings = [r.text for r in ratings_tags]
print(len(ratings))
print(ratings)
```

```python
jobs = driver.find_elements_by_xpath("//li[@class='react-job-listing css-wp148e eigr9kq3']")
print(len(jobs))

company_name = []
days_posted_ago = []
ratings = []

for j in jobs:

    # Scrapping for company name from the page
    company_name_tags = j.find_elements_by_xpath(".//div[@class='d-flex justify-content-between align-items-start']")
    if (len(j.find_elements_by_xpath(".//div[@class='d-flex justify-content-between align-items-start']")) != 0):
        for c in company_name_tags:
            company_name.append(c.text)
    else:
        company_name.append(None)
    print(company_name)

    # Scrapping for no of days posted before
    days_posted_ago_tags = j.find_elements_by_xpath(".//div[@class='d-flex align-items-end pl-std css-mi55ob']")
    if(len(j.find_elements_by_xpath(".//div[@class='d-flex align-items-end pl-std css-mi55ob']")) !=0):
        for d in days_posted_ago_tags:
            days_posted_ago.append(d.text)
    else:
        days_posted_ago.append(None)
    print(days_posted_ago)
    
    # Scrapping for the ratings
    ratings_tags = j.find_elements_by_xpath(".//div[@class='d-flex flex-column css-x75kgh e1rrn5ka3']//span[@class='css-19pjha7 e1cjmv6j1']")
    if(len(j.find_elements_by_xpath(".//div[@class='d-flex flex-column css-x75kgh e1rrn5ka3']//span[@class='css-19pjha7 e1cjmv6j1']")) != 0):
        for r in ratings_tags:
            ratings.append(r.text)
    else:
        ratings.append(None)
    print(ratings)
```

```python
print(len(company_name))
print(company_name)
print(len(days_posted_ago))
print(days_posted_ago)
print(len(ratings))
print(ratings)
```

# Question 6:-
    : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
    1. Brand
    2. Product Description
    3. Price
    4. Discount %

```python
# Writing a request to fetch the data from the web page 
driver.get("https://www.flipkart.com/")
```

```python
# Clicking the cross button which asks us to sign in
cross_button = driver.find_element_by_xpath("//div[@class='_2QfC02']//button[@class='_2KpZ6l _2doB4z']")
cross_button.click()
```

```python
# Finding  element for the search for products, brands and more search bar
search_bar = driver.find_element_by_xpath("//input[@title='Search for products, brands and more']")
search_bar.send_keys('sunglasses')
```

```python
# Finding the element for clicking the search botton
search_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
# Finding the elements for the brand name and scrapping the text
Brand_tags = driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
Brand = [b.text for b in Brand_tags]

# Finding the elements for the product description and scrapping the text
Product_Description_tags = driver.find_elements_by_xpath("//a[@class='IRpwTa']")
Product_Description = [pd.text for pd in Product_Description_tags]

# Finding the elements for the price and scrapping the text
Price_tags = driver.find_elements_by_xpath("//div[@class='_30jeq3']")
Price = [p.text for p in Price_tags]

# Finding the elements for the discount and scrapping the text
Discount_tags = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
Discount = [d.text.strip(' off') for d in Discount_tags]
```

```python
print(len(Brand))
print(Brand)
print(len(Product_Description))
print(Product_Description)
print(len(Price))
print(Price)
print(len(Discount))
print(Discount)
```

```python
# Creating empty lists 
Brand = []
Product_Description = []
Price = []
Discount = []

while (len(Brand) and len(Product_Description) and len(Price) and len(Discount)) < 100:
    
    # Scrapping the information of the whole page
    sunglasses_page = driver.find_elements_by_xpath("//div[@class='_1xHGtK _373qXS']//div[@class='_2B099V']")
    print(len(sunglasses_page))
    
    for s in sunglasses_page:
        
        while (len(Brand) and len(Product_Description) and len(Price) and len(Discount)) < 100:
    
            # Scrapping the brand name from the page
            Brand_tags = s.find_elements_by_xpath(".//div[@class='_2WkVRV']")
            if (len(s.find_elements_by_xpath(".//div[@class='_2WkVRV']")) != 0):
                for b in Brand_tags:
                    Brand.append(b.text)
            else:
                Brand.append(None)
            #print(Brand)

            # Scrapping the production description from the page
            Product_Description_tags = s.find_elements_by_xpath(".//a[@class='IRpwTa' or @class='IRpwTa' or @class='IRpwTa _2-ICcC']")
            if (len(s.find_elements_by_xpath(".//a[@class='IRpwTa' or @class='IRpwTa']")) != 0):
                for pd in Product_Description_tags:
                    Product_Description.append(pd.text)
            else:
                Product_Description.append(None)
            #print(Product_Description)

            # Scrapping the price from the page
            Price_tags = s.find_elements_by_xpath(".//div[@class='_30jeq3']")
            if (len(s.find_elements_by_xpath(".//div[@class='_30jeq3']")) != 0):
                for p in Price_tags:
                    Price.append(p.text)
            else:
                Price.append(None)
            #print(Price)

            # Scrapping the discount from the page
            Discount_tags = s.find_elements_by_xpath(".//div[@class='_3Ay6Sb']")
            if (len(s.find_elements_by_xpath(".//div[@class='_3Ay6Sb']")) != 0):
                for d in Discount_tags:
                    Discount.append(d.text.strip(' off'))
            else:
                Discount.append(None)
            #print(Discount)
        
            break
        
    # Clicking the next button
    next_button = driver.find_element_by_xpath("//a[@class='_1LKTO3']/span[contains(text(),'Next')]")
    next_button.click()

    time.sleep(5)
```

```python
print(len(Brand))
print(Brand)
print(len(Product_Description))
print(Product_Description)
print(len(Price))
print(Price)
print(len(Discount))
print(Discount)
```

```python
import pandas as pd
#Building the data frame
sunglasses_data = pd.DataFrame({
        'Brand' : Brand,
        'Product_Description' : Product_Description,
        'Price' : Price,
        'Discount' : Discount
})
    
print('First 100 sunglasses. ----->')
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', None)
sunglasses_data
```

    2. Review_summary

    You have to scrape this data for first 100 reviews.

```python
# Writing a request to fetch the data from the web page 
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace")
```

```python
# Clicking to get all the reviews 
reviews_button = driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']/span")
reviews_button.click()
```

```python
Rating_stars = []
Review_summary = []
Full_review = []

while (len(Rating_stars) and len(Review_summary) and len(Full_review)) < 100:

    # Finding the review tag which contains the reviews of all the three categories mentioned above
    reviews_page = driver.find_elements_by_xpath("//div[@class='_27M-vq']")
    print(len(reviews_page))

    for rp in reviews_page:
    
        #while (len(Rating_stars) and len(Review_summary) and len(Full_review)) < 100:
    
        # Finding the tag for the stars and then scrapping it
        Rating_stars_tags = rp.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq'] | .//div[@class='_3LWZlK _1rdVr6 _1BLPMq'] | .//p[@class='_2-N8zT']")
        Rating_stars.append(Rating_stars_tags.text)
                     
        # Finding the tag for the summary of reviews and then scrapping it
        Review_summary_tags = rp.find_element_by_xpath(".//p[@class='_2-N8zT']")
        Review_summary.append(Review_summary_tags.text)
    
        # Finding the tag for full reviews and then scrapping it
        Full_review_tags = rp.find_element_by_xpath(".//div[@class='t-ZTKy']/div/div")
        # First we need to run a loop where it clicks on all the read more reviews and then extract the text from it
        read_more_button = driver.find_elements_by_xpath("//div[@class='t-ZTKy']//span[@class='_1BWGvX']")
        for r in read_more_button:
            try:
                r.click()
                time.sleep(2)
            except:
                print(None)
            break 
    
        Full_review.append(Full_review_tags.text)
        
    # Clicking the next button
    next_button = driver.find_element_by_xpath("//a[@class='_1LKTO3']/span[contains(text(),'Next')]")
    next_button.click()

    time.sleep(5)
```

```python
Brand_name = []
Short_desc = []
Price = []

while (len(Brand_name) and len(Short_desc) and len(Price)) < 100:

    # Finding all the tags for the products
    shoes_page = driver.find_elements_by_xpath("//li[@class='product-base']//div[@class='product-productMetaInfo']")
    print(len(shoes_page))
    
    for s in shoes_page:
        
        while (len(Brand_name) and len(Short_desc) and len(Price)) < 100:

            # Finding the tags for the brand name and then scrapping the data
            Brand_name_tags = s.find_element_by_xpath(".//h3[@class='product-brand']")
            Brand_name.append(Brand_name_tags.text)
    
            # Finding the tags for the short product description and then scrapping the data
            Short_desc_tags = s.find_element_by_xpath(".//h4[@class='product-product']")
            Short_desc.append(Short_desc_tags.text)
    
            # Finding the tags for the price and then scrapping the data
            try:
                Price_tags = s.find_element_by_xpath(".//div[@class='product-price']//span[@class='product-discountedPrice']")
                Price.append(Price_tags.text.strip(" /nRs. /n/n"))
            except:
                Price_tags = s.find_element_by_xpath(".//div[@class='product-price']/span")
                Price.append(Price_tags.text.strip(" /nRs. /n/n"))
        
            break
        
    # Clicking the next button
    next_button = driver.find_element_by_xpath("//li[@class='pagination-next']/a[contains(text(),'Next')]")
    next_button.click()

    time.sleep(5)
```

# Question 10 :-
    Go to webpage https://www.amazon.in/ .Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” and “Intel Core i9”. 
    After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
    1. title
    2. Ratings
    3. Price

```python
# Writing a request to fetch the data from the web page 
driver.get("https://www.amazon.in/")
```

```python
# Finding  element for the search bar
search_bar = driver.find_element_by_xpath("//input[@type='text']")
search_bar.send_keys('Laptop')
```

```python
# Finding the element for clicking the search botton
search_button = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_button.click()
```

```python
# Finding the elements for the CPU type filter
i7_box = driver.find_element_by_xpath("//li[@aria-label='Intel Core i7']//div[@class='a-checkbox a-checkbox-fancy s-navigation-checkbox aok-float-left']")
i7_box.click()

time.sleep(5)

i9_box = driver.find_element_by_xpath("//li[@aria-label='Intel Core i9']//div[@class='a-checkbox a-checkbox-fancy s-navigation-checkbox aok-float-left']")
i9_box.click()
```

```python
Title = []
Ratings = []
Price = []
        
while (len(Title) and len(Ratings) and len(Price)) < 10:

    # Finding the tags for the brand name and then scrapping the data
    Title_tags = driver.find_elements_by_xpath("//div[@class='sg-col-inner']//span[@class='a-size-medium a-color-base a-text-normal']")
    for t in Title_tags:
        while (len(Title)) < 10:
            Title.append(t.text)
            break
    
    # Finding the tags for the short product description and then scrapping the data
    Ratings_tags = driver.find_elements_by_xpath("//div[@class='sg-col-inner']//div[@class='a-row a-size-small']/span")
    for r in Ratings_tags[0:len(Ratings_tags):2]:
        while (len(Ratings)) < 10:
            Ratings.append(r.get_attribute('aria-label'))
            break
    
    # Finding the tags for the price and then scrapping the data
    Price_tags = driver.find_elements_by_xpath("//div[@class='sg-col-inner']//span[@aria-hidden='true']//span[@class='a-price-whole']")
    for p in Price_tags:
        while (len(Price)) < 10:
            Price.append(p.text)
            break
```

```python
print(len(Title))
print(Title)
print(len(Ratings))
print(Ratings)
print(len(Price))
print(Price)
```

```python
import pandas as pd
#Building the data frame
laptops_data = pd.DataFrame({
        'Title' : Title,
        'Ratings ' : Ratings,
        'Price(₹)' : Price
})
    
print('First 10 laptops. ----->')
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', None)
laptops_data
```
